#### 1. What is the concept of cyclical momentum ?

**Answer**
The concept of cyclical momentum involves dynamically adjusting the momentum term during training cycles to control the update velocity of the optimization algorithm, potentially helping to escape local minima and converge more effectively.

#### 2. What callback keeps track of hyperparameter values (along with other data) during training ?

**Answer**
The Recorder callback keeps track of hyperparameter values and other data during training.

#### 3. In the color dim plot, what does one column of pixels represent ?

**Answer**
In the color dim plot, one column of pixels represents the intensity values of a specific color channel across the entire image's vertical dimension.

#### 4. In color dim, what does "poor teaching" look like? What is the reason for this ?

**Answer**
"Poor teaching" in the color dim plot refers to columns of pixels that are very similar in color across the image, leading to a lack of informative variation. This can hinder the model's ability to learn distinct features.

#### 5. Does a batch normalization layer have any trainable parameters ?

**Answer**
Yes, a batch normalization layer has trainable parameters: scale and shift (also known as gamma and beta), which allow the layer to learn the optimal mean and variance for each batch during training.

#### 6. In batch normalization during preparation, what statistics are used to normalize? What about during the validation process ?

**Answer**During batch normalization, mean and variance statistics calculated within each batch are used for normalization during training. During validation, population statistics (computed over the entire training dataset) are used to maintain consistency and avoid data leakage.

#### 7. Why do batch normalization layers help models generalize better ?

**Answer**Batch normalization layers help models generalize better by reducing internal covariate shift, stabilizing training, and enabling the network to learn more robust features by maintaining consistent activations.

#### 8.Explain between MAX POOLING and AVERAGE POOLING is number eight ?

**Answer**Max pooling retains the maximum value within a region, emphasizing prominent features. Average pooling calculates the average value, providing a smoother representation of features.

#### 9. What is the purpose of the POOLING LAYER ?

**Ans:** Pooling layers are used to reduce the dimensions of the feature maps. Thus, it reduces the number of parameters to learn and the amount of computation performed in the network.

The pooling layer summarises the features present in a region of the feature map generated by a convolution layer.

#### 10. Why do we end up with Completely CONNECTED LAYERS ?

**Answer**Completely connected (fully connected) layers combine features learned from previous layers and produce final predictions. They consolidate information for classification or regression tasks.

#### 11. What do you mean by PARAMETERS ?

**Answer**Parameters refer to the learnable variables within a neural network that are optimized during training to influence the network's output. They include weights and biases.

#### 12. What formulas are used to measure these PARAMETERS?

**Answer**The formulas used to measure parameters include:

Number of parameters in a layer: (Number of input units + 1) * Number of output units
Total number of parameters in a network: Sum of parameters across all layers
